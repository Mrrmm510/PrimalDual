### try the implemented total variation model

load a video

```python
cap = cv2.VideoCapture('../data/rabbit.mov')

video = list()
while(cap.isOpened()):
    ret, frame = cap.read()
    if not ret:
        break
    video.append(cv2.resize(cv2.cvtColor(frame[400:1000, 150:-150], cv2.COLOR_BGR2GRAY), None, fx=0.25, fy=0.25))
fps = cap.get(cv2.CAP_PROP_FPS)
cap.release()

video = np.array(video, dtype='float32') / 255.
coef = np.array([1, -1])
lambd = 0.5
```

```python
print(video.shape, np.prod(video.shape))
```

```python
%%time
tv = TotalVariation3D(lambd = lambd, param_w = 1., param_h = 1., param_d = 0., saturation = True, extended_output = True, max_iter = 100, coef = coef)
smoothed, obj = tv.transform(video)
```

objective function

```python
plt.plot(obj[:])
plt.grid(True)
```

result

```python
plt.figure(figsize=(20, 10))
plt.subplot(1,3,1)
plt.imshow(video[1], cmap='gray')
plt.grid(False)
plt.title('Original')
plt.subplot(1,3,2)
plt.imshow(smoothed[1], cmap='gray')
plt.grid(False)
plt.title('Smoothed')
plt.subplot(1,3,3)
plt.imshow(np.abs(video[1] - smoothed[1]), cmap='gray')
plt.grid(False)
plt.title('Diff')
plt.show()
```

```python
plt.figure(figsize=(20, 10))
plt.plot(video[0, 100, :], 'r')
plt.plot(smoothed[0, 100, :], 'b')
plt.grid()
```

```python
fourcc = cv2.VideoWriter_fourcc(*'MP4V')
out = cv2.VideoWriter('../data/smoothed_rabbit.mp4',fourcc, fps, (224, 150), isColor=False)

for frame1, frame2 in zip(video, smoothed):
    frame = np.hstack([frame1 * 255, np.clip(frame2 * 255, 0, 255)])
    out.write(np.uint8(frame))
out.release()
```
